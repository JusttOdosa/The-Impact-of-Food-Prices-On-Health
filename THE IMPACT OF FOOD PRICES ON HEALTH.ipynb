# 5. Examining The Affordability of an Healthy Diet

```python
analysis_df = us_share_of_pop_that_cannot_afford_healthy_diet.merge(us_daily_cost_of_healthy_diet_per_person, on=['Entity', 'Year'], how='inner')
analysis_df = analysis_df.merge(us_share_of_pop_that_is_undernourished, on=['Entity', 'Year'], how='left')

# Merging relevant datasets
```

```python
analysis_df
```

```python
correlation_vars = [
        'Share of the population who cannot afford a healthy diet',
        'Cost of a healthy diet'
    ]
correlations = analysis_df[correlation_vars].corr()

correlations

# Calculating correlations
```

```python
plt.subplot(2, 2, 1)
sns.heatmap(analysis_df[correlation_vars].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Diet Affordability Metrics')

# Correlation heatmap
```

```python
 regional_summary = analysis_df.groupby('Entity').agg({
        'Share of the population who cannot afford a healthy diet': 'mean',
        'Cost of a healthy diet': 'mean',
        'Prevalence of undernourishment (%)': 'mean'
    }).sort_values('Share of the population who cannot afford a healthy diet', ascending=False)
    
regional_summary
    
# Regional Analysis
```
